```python
# (type)은 parameter
def single_or_multi(type):
    if '/'in type:
        return 'Multi'
    return 'Single'
```

```python
## pandas.Series.apply##
# Series.apply(func, convert_dtype=True, args=(), **kwargs)
# axis가 0 or 'index'인 경우 행마다 함수를 적용.
# axis가 1 or ‘columns’인 경우 열마다 함수를 적용.

pokemons.apply(single_or_multi).value_counts()
```

```python
bins = [1, 20, 30, 50, 70, 100]
labels = ["미성년자", "청년", "장년", "중년", "노년"]
titanic_age = pd.cut(titanic.age,bins=bins,labels=labels)
titanic_rate = titanic_age.value_counts(normalize = True)           #나이 카테고리별 비율 계산
titanic_rate.sum()
```
