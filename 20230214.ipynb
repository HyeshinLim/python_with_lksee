```python
if not os.path.isdir("./female"):
    os.mkdir("./female")
if not os.path.isdir("./male"):
    os.mkdir("./male")
# enumerate -> list에 번호를 생성 
female = list()
male = []
for idx, file in enumerate(os.listdir("./이름_생년_성별_10000/")):
    #if idx == 10 : break
    f = open(f"./이름_생년_성별_10000/{file}", "r", encoding='utf-8')
    tmp = f.read()
    f.close()
    if file.split("_")[2][0] in ['1', '3']:
        shutil.move(f"./이름_생년_성별_10000/{file}", "./male/")
        male.append(tmp)
    else:
        shutil.move(f"./이름_생년_성별_10000/{file}", "./female/")
        female.append(tmp)
```
